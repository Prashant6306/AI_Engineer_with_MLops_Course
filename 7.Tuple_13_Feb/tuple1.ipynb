### Cloning

```python
tuple1=(10,20,30)
```

```python
original_tuple = (10, 20, 30)
cloned_tuple = original_tuple.copy()

print(id(original_tuple))  # (10,20,30) address: 100123
print(id(cloned_tuple))    # (10,20,30) address: 100124
# Same content, different memory!
```

### Searching — 'in' Membership Operator

```python
nums = (1, 2, 3, 4, 5)
print(3 in nums)     # True
print(8 in nums)     # False
```

### Identity — 'is' Operator

```python
p = (1, 2, 3)
q = p             # aliasing — same object
r=(10,20,30)
print(p is q)     # True  (same memory)
print(q is p) 
```

```python
print(r is p)
```

### Comparison

```python
a = (1, 2, 3, 4)
b = (1, 2, 5)
print(a < b)    # True   (3 < 5 at index 2)
print(a > b)    # False
print(a == b)   # False
```

## Emptiness

```python
my_tuple = ()

if not my_tuple:
    print("tuple is empty")
else:
    print("tuple is not empty")

# Output: tuple is empty
```

```python
my_tuple = (10,)

if not my_tuple:
    print("tuple is empty")
else:
    print("tuple is not empty")

# Output: tuple is empty
```
