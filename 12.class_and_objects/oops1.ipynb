```python
class Vehical:
    def __init__(self):
        self.color="red"
        self.top_speed=150
        self.milage=20
    def display_car_details(self):
        print("car colour is",self.color)
        print("car's milage is",self.milage)
        print("car's top speed is",self.top_speed)

obj1=Vehical()
obj1.display_car_details()
```
